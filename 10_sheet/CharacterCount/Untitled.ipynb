```python
char_occurences = {}
sum_count = 0
with open('part-r-00000') as f:
    for line in f:
        char, count = line.split()
        count = int(count)
        char_occurences[char] = count
        sum_count += count
```

```python
for key, val in char_occurences.items():
    char_occurences[key] /= sum_count 
```

```python
sorted_occurences = sorted(char_occurences.items(), key=lambda x: -x[1])
```

```python
plt.figure(figsize=(20,8), dpi=120)
chars, occurences = list(zip(*sorted_occurences))
plt.bar(chars, occurences)
```
